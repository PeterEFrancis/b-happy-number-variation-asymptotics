# b-happy number variation asymptotics

```python
from math import log
import matplotlib.pyplot as plt
from numba import njit
import numpy as np

@njit
def sum_d(n):
    if n < 10:
        return n
    else:
        return (n % 10) +  sum_d(n // 10)

@njit
def compute(bound1, bound2, mod):
    count = 0
    data = np.zeros(bound2 - bound1)
    last = 0
    for i in range(bound1 + 1, bound2):
        if i % sum_d(i) == mod:
            count += 1
        if i % 100 == 0 or i < 1000:
            last = (count / (i / log(i)))
        data[i-bound1] = last
    return data
```

```python
fig, ax = plt.subplots(figsize=(35,15))

for mod in range(100):
    data = compute(mod + 1, 1_500_000, mod)
    ax.plot(data);
    print(f"mod {mod} -- > {data[-1]}")
```

```python
mod_to_cst = []

for mod in range(100):
    data = compute(mod + 1, 1_000_000, mod)
    mod_to_cst.append(data[-1])
```

```python
fig, ax = plt.subplots(figsize=(35,15))
ax.plot(mod_to_cst);
```
